```python
qnt= [0,0,0,0]

for i in range(df['Cluster'].max() + 1):
    dfC = df.loc[df['Cluster'] == i]
    qnt[i] = dfC.shape[0]
```

```python
clusters = [0,1,2,3]
clusterDF = df['Cluster']
raised = [0,0,0,0]
raisedRelativo = [0,0,0,0]
raisedDF = df['Raised']
for i in range(6541):
    raised[clusterDF.loc[i]] += raisedDF.loc[i]
for i in range(4):
    raisedRelativo[i] = raised[i]/qnt[i]
```

```python
import plotly.express as px
raisedClusters = {'Raised Total':raised, 'Clusters': clusters, 'Qnt':qnt, 'Raised Relativo':raisedRelativo}
dfRaisedClusters = pd.DataFrame(data=raisedClusters) 

# Use column names of df for the different parameters x, y, color, ...
fig = px.bar(dfRaisedClusters,x='Clusters', y='Raised Relativo', color = 'Clusters',
                 title="Média Raised Clusters"
                )

fig.show()
```

```python
# Use column names of df for the different parameters x, y, color, ...
fig = px.bar(dfRaisedClusters,x='Clusters', y='Raised Total', color = 'Clusters',
                 title="Raised Total por Cluster"
                )

fig.show()
```
